```python
import pandas as pd
import numpy as np
```

```python
from cnn_class.cnn import CNN
from cnn_class.utils import *
```

```python
np.random.seed(21)
size = (32,32)
n_samples = 1000
```

```python
data = np.random.rand(n_samples,*size)*0.2
data = data.reshape(n_samples,1,*size)
data
```

```python
labels = np.random.randint(0,2,size=n_samples)
labels
```

```python
def preprocess(x):
    mean = np.mean(x,axis=0)
    std = np.std(x,axis=0)

    return ((x - mean)/std)
```

```python
data_tmp = data.reshape(-1,size[0]*size[1])
data_pro = preprocess(data_tmp)
data = data_tmp.reshape(n_samples,1,*size)
```

```python
split_inds = np.random.permutation(n_samples)
data_train, data_val, data_test = data[split_inds[:700]], data[split_inds[700:800]], data[split_inds[800:]]
labels_train, labels_val, labels_test = labels[split_inds[:700]], labels[split_inds[700:800]], labels[split_inds[800:]]
```

```python
cnn = CNN()
```

```python
cnn.train(
    train_data=data_train,
    train_labels=labels_train,
    val_data=data_val,
    val_labels=labels_val,
    num_epochs=1000,
    learning_rate=5e-3,
    validate=True,
    regularization=5e-4,
    verbose=0
)
```

```python
cnn.predict(
    X=data_test,
    y=labels_test,
    regularization=5e-4,
    verbose=0
)
```
